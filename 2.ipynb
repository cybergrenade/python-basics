```python
class Theatre:
  def __init__(self, name):
    self.name = name

  def add_performance(self, pname, start):
    return Performance(self.name, pname, start, self)

class Performance(Theatre):
  def __init__(self, name, pname, start, theatre):
    if not isinstance(theatre, Theatre):
      raise ValueError("Performance cannot exist without Theatre")
    super().__init__(name)
    self.pname = pname
    self.start = start

  def display(self):
    print(f"Theatre: {self.name}, performance {self.pname} at {self.start}")


bolshoi = Theatre("Bolshoi Theatre")


b = bolshoi.add_performance("Shchelkunchik", "21:00")
b.display()

```

```python
class Journey:
    def __init__(self, name):
        self.name = name
        self.attractions = []

    def add_attraction(self, attraction):
        if not isinstance(attraction, Attraction):
            raise ValueError("Only Attraction objects can be added")
        self.attractions.append(attraction)

    def display(self):
        for a in self.attractions:
            print(f"{a.name}: {a.description}")

class Attraction(Journey):
    def __init__(self, name, description, journey):
        super().__init__(journey.name)
        self.name = name
        self.description = description
        journey.add_attraction(self)

eurotrip = Journey("Евротур")
colosseum = Attraction("Колизей", "Амфитеатр в Риме", eurotrip)
eiffel_tower = Attraction("Эйфелева башня", "Символ Парижа", eurotrip)

eurotrip.display()
```

```python
from abc import ABC, abstractmethod
class Bank(ABC):
  @abstractmethod
  def calculateInterest(self, amount):
    pass

class SavingsBank(Bank):
  def calculateInterest(self, amount):
     return amount * 0.18

class InvestmentBank(Bank):
  def calculateInterest(self, amount):
     return amount * 0.24

a = SavingsBank()
b = InvestmentBank()

print(a.calculateInterest(100000), b.calculateInterest(100000))
```
