```python
def frequency(text):
    words = text.lower().split()
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    return freq

print(frequency("aaa bbb aaa aaa bbb"))
```

```python
def invert(d):
    inverted = {}
    for k, v in d.items():
        if v in inverted:
            if isinstance(inverted[v], list):
                inverted[v].append(k)
            else:
                inverted[v] = [inverted[v], k]
        else:
            inverted[v] = k
    return inverted

print(invert({'a': 1, 'b': 1, 'c': 2}))
```

```python
def filter(d, threshold):
    return {k: v for k, v in d.items() if v > threshold}

print(filter({'a': 10, 'b': 5, 'c': 15}, 7))
```

```python
def remove_keys(d, keys):
    for key in keys:
        d.pop(key, None)
    return d

print(remove_keys({'a': 1, 'b': 2, 'c': 3}, ['a', 'c']))
```

```python
def intersect(d1, d2):
    return {k: d1[k] for k in d1 if k in d2 and d1[k] == d2[k]}

print(intersect({'a': 1, 'b': 2}, {'a': 1, 'c': 3}))
```

```python
def average(d):
    values = [v for v in d.values() if isinstance(v, (int, float))]
    return sum(values) / len(values) if values else 0

print(average({'a': 1, 'b': 2, 'c': 3}))
```

```python
def group_by_type(t):
    result = {}
    for item in t:
        type_key = type(item).__name__
        result.setdefault(type_key, []).append(item)
    return result

print(group_by_type((1, "hello", 2.5, 2)))
```
