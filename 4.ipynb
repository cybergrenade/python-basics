```python
def filter_palindromes(s):
    words = s.split()
    last_word = words[-1]
    return [word for word in words[:-1] if word != last_word and word == word[::-1]]


s = "racecar level hello tenet level"
result = filter_palindromes(s)
print(result)
```

```python
def replace_count(s):
  k = s.count(':')
  s = s.replace(':', ';')
  return s, k

s = 'a:aa;aa:a:aA:'
print(replace_count(s))
```

```python
def c(s):
    words = s.rstrip('.').split(',')

    return sum(1 for word in words if word.strip().lower().startswith('a'))

s = "apple,banana,ant,avocado,cat."
result = c(s)
print(result)
```

```python
#input: 1.5 2.3 10.0 14.2
with open('a.txt', 'r') as f:
  n = [float(i)**2 for i in f.read().split()]

with open('a.txt', 'w') as f:
  f.write(' '.join(str(i) for i in n))

with open('a.txt', 'r') as f:
  print(f.read())
```
