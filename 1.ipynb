```python
def horse(a, b, e, f):
  return abs(a-e) == 2 and abs(b-f) == 1 or abs(a-e) == 1 and abs(b-f) == 2

def A(a, b, c, d, e, f):
  if not(horse(a, b, e, f)) or e == c or f == d:
    return False

  return True

def B(a, b, c, d, e, f):
  if not(horse(a, b, e, f)) or e == c or f == d or abs(e-c) == abs(f-d):
    return False

  return True

def C(a, b, c, d, e, f):
  if not(horse(a, b, e, f)) or abs(e-c) == abs(f-d):
    return False

  return True
a = int(input('Enter a: '))
b = int(input('Enter b: '))
c = int(input('Enter c: '))
d = int(input('Enter d: '))
e = int(input('Enter e: '))
f = int(input('Enter f: '))

print('a) yes' if A(a, b, c, d, e, f) else 'a) no')
print('b) yes' if B(a, b, c, d, e, f) else 'b) no')
print('c) yes' if C(a, b, c, d, e, f) else 'c) no')
```

```python
B = list(map(int, input("Введите элементы массива через пробел: ").split()))
if B[0] < B[1]:
  max = B[0]
else:
  max = -10**10
N = len(B)
for i in range (1,N-1):
  if B[i] < B[i-1] and B[i] < B[i+1] and B[i] > max:
    max = B[i]
if B[N-1] < B[N-2] and B[N-1] > max:
  max = B[N-1]
print(max)
```
