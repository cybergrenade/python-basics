```python
def calc_t(a, b, c):
  return (a*c-a*b)/(b*b-c*c) + (b*b+a*b)/(3*a*a-3*a*b)

a = int(input("enter a: "))
b = int(input("enter b: "))
c = int(input("enter c: "))
t = calc_t(a, b, c)
print(f't = {t}')
```

```python
from math import sin, exp
def calc_w(x):
  return sin(2*x)**3*exp(-x)

x = int(input("enter x: "))
w = calc_w(x)
print(f'w = {w}')
```

```python
from math import floor
def cats(x, y):
  speed = 1.5/1.5/1.5
  return speed*x*y

x = int(input("how many cats?: "))
y = int(input("how many hours?: "))
a = floor(cats(x, y))
print(f'{x} cats eat {a} mice in {y} hours')
```

```python
def f(a, b):
  return a%2 != 0 or b%2 != 0

a = int(input("enter a: "))
b = int(input("enter b: "))
print("Yes" if f(a, b) else "No")
```
